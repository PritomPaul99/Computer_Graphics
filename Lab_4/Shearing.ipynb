# 2D Transformation

## Shearing

```python
import matplotlib.pyplot as plt
import numpy as np
import math
import cv2
```

```python
image = cv2.imread("letter-p_12785849.png", 0)
h, w = image.shape

newImage = np.zeros((h + 250, w + 250), dtype="uint8")

rad = math.radians(30)

for i in range (h):
    for j in range (w):
        pix = image[i][j]
        
        xf = i + j * math.tan(rad)
        yf = j
        
        x2 = int(xf)
        y2 = int(yf + 100)
        
        if (0 <= x2 < h + 250 and 0 <= y2 < w + 250):
            newImage[x2, y2] = pix
```

```python
plt.subplot(1, 2, 1)
plt.imshow(image)

plt.subplot(1, 2, 2)
plt.imshow(newImage)

plt.tight_layout()

plt.show()
```
