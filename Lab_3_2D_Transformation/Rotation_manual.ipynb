## Rotation (Linear, without function)
> Not recommended, use function.

```python
import matplotlib.image as img
import matplotlib.pyplot as plt
import numpy as np
import math
import cv2
```

```python
image = cv2.imread("cube_3x3.PNG", 0)

h, w = image.shape[:2]

print(h, w)
```

```python
newImg =  np.zeros((w+250, h+250), dtype = "uint8")

print(newImg, newImg.shape)
```

```python
angle = int(input())

rad = math.radians(angle) # Degree to radian

print("Angle:", angle)
print("Radian:", rad)
```

```python
for i in range (h):
    for j in range (w):
        pix = image[i][j]
        pnt = np.array([i, j]).reshape(-1, 1)
        rotation_mat = np.array([math.cos(rad), -math.sin(rad), math.sin(rad), math.cos(rad)])
        rotation_mat = rotation_mat.reshape(2, 2)
        newMat = rotation_mat @ pnt # Matrix multiplication
        xf = newMat[0][0]
        yf = newMat[1][0]
        x2 = int(xf + 150)
        y2 = int(yf)
        newImg[x2, y2] = pix
```

```python
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.subplot(1, 2, 2)
plt.imshow(newImg)
plt.tight_layout()
plt.show()
```
